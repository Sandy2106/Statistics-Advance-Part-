# Statistics Advance Part 1

1. What is a random variable in probability theory?
- A random variable is a numerical outcome of a random experiment. It assigns numerical values to each outcome in the sample space.

2. What are the types of random variables?
-
 - Discrete Random Variable: Takes countable values (e.g., number of heads in coin flips).
 - Continuous Random Variable: Takes an infinite number of possible values within an interval (e.g., height, temperature).

3. What is the difference between discrete and continuous distributions?
-
 - Discrete Distribution: Defined for countable values (e.g., binomial, Poisson).
 - Continuous Distribution: Defined over an interval, requiring integration to find probabilities (e.g., normal, uniform).

4. What are probability distribution functions (PDF)?
-
 - For discrete variables: The function assigns probabilities to specific values.
 - For continuous variables: The probability density function (pdf) gives the relative likelihood of a value but requires integration to compute actual probabilities.

5. How do cumulative distribution functions (CDF) differ from probability
  distribution functions (PDF)?
-  
 - PDF (Probability Density Function): Represents the likelihood of a value occurring (for continuous distributions).

 - CDF (Cumulative Distribution Function): Gives the probability that the random variable is less than or equal to a certain value.

6. What is a discrete uniform distribution?
-  A distribution where all possible outcomes have the same probability (e.g., rolling a fair die).

13. What is the Central Limit Theorem (CLT), and why is it critical in
   statistics?
- States that the sampling distribution of the sample mean approaches a normal
 distribution as sample size increases, regardless of the original population distribution.

14. How does the Central Limit Theorem relate to the normal distribution?
- The CLT justifies using normal approximations in probability and statistics, allowing inference even when the original data isn't normal.

15. What is the application of Z statistics in hypothesis testing?
- Used in hypothesis testing to determine how far a sample statistic is from the population mean in standard deviation units.
- Helps in testing means and proportions when population variance is known.

17. What are point estimates and interval estimates in statistics?
- Point Estimate: A single value used to estimate a parameter (e.g., sample mean).
- Interval Estimate: A range of values (e.g., confidence intervals) estimating the parameter with a certain confidence level.

18. What is the significance of confidence intervals in statistical analysis?
- Significance of Confidence Intervals: Provides a range of values within which the population parameter is likely to lie, with a specified confidence level (e.g., 95%).

19. What is the relationship between a Z-score and a confidence interval?
- Z-score determines the critical value for constructing confidence intervals.

20. How are Z-scores used to compare different distributions?
- Standardizes different distributions by converting values into a common scale.

- Useful for comparing different datasets with different means and standard deviations.
